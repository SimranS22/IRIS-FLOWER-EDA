#### 3.2.2. Fill or Drop missing values

Here, we will be dealing with only with _IRIS1.CSV_

**Filling missing values**

The values can be mean, median, mode or any value.

_sepal_length(mean)_

````python
iris1df['sepal_length'].fillna(value=iris1df['sepal_length'].mean(), inplace = True) 
````

````python
iris1df.isnull().sum()
````

_sepal_width(mode)_

````python
iris1df['sepal_width'].fillna(value=iris1df['sepal_width'].mode().values[0], inplace = True) 
````

````python
iris1df.isnull().sum()
````

_petal_length(median)_

````python
iris1df['petal_length'].fillna(value=iris1df['petal_length'].median(), inplace = True) 
````

````python
iris1df.isnull().sum()
````

_petal_width(place a value)_

````python
iris1df['petal_width'].fillna(value=0.1, inplace = True) 
````

````python
iris1df.isnull().sum()
````

**Dropping missing values**

Now for species, we cannot fill the missing values with arbitrary data.

````python
iris1df = iris1df.dropna(axis = 0, how ='any')
````

````python
print(iris1df.isnull().sum())
````
